```python
fifa = pd.read_csv('/content/fifa_data.xls')
fifa
```

```python
fifa['Weight']
```

```python
def stripdata(x):
    if type(x)==str:
        return float(x[0:3])
    else:
        return x
```

```python
# def stripdata(x):
#     if type(x)==str:
#         return int(x.strip('lbs'))
#     else:
#         return x
```

```python
stripdata('159lbs')
```

```python
fifa['Weight'] = fifa['Weight'].apply(stripdata)
```

```python
fifa['Weight']
```

```python
light = fifa[fifa.Weight < 125].count()[0]
light
```

```python
light = fifa[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
weights
```

```python
plt.figure(figsize=(6,6))

#print(weights)

label = ['under 125', '125-150', '150-175', '175-200', 'over 200']

explode = [.4,.2,0,0,.4]

plt.pie(weights, labels=label, explode=explode, autopct='%.2f %%',  pctdistance=0.8)

#plt.pie(weights, labels=label, autopct='%.2f %%',  pctdistance=0.7)

plt.legend(title="Weight Range")

plt.show()
```
