##Distribution Plots

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
# Loading data set
tips = sns.load_dataset('tips')
tips
```

```python
# Lets get a concise summary of the data set.
tips.info()
```

So, our data have 244 entries and 7 columns. Let's check the head of the data!

```python
tips.tip.hist(bins=30)
```

##Distplot()

```python
# lets check how the total bill are distributed in the data set
sns.distplot(tips['tip'], bins=20)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## Jointplot()

```python
tips.plot.scatter(x='total_bill', y='tip')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, hue='sex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```
