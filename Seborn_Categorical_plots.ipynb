# seaborn: statistical data visualization
## Plotting with categorical data

```python
# import seaborn library and load the dataset 'tips'
import seaborn as sns
tips = sns.load_dataset('tips')
tips
```

```python
tips['day'].unique()
```

```python
tips[tips['day']=='Thur']
```

```python
# once again checking the head of the dataset
tips
```

### Categorical scatterplots:
###stripplot()

```python
sns.stripplot(x="day", y="total_bill",data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", jitter=False, data=tips)
```

```python
sns.stripplot(x="day", y="total_bill",data=tips,hue="sex")#, palette='coolwarm')
```

###swarmplot()

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="smoker",palette='coolwarm')
```

```python
sns.swarmplot(x="day",y="total_bill",  data=tips);
```

```python
sns.swarmplot(x="day",y="total_bill",  hue="time", data=tips);
```

###boxplot()

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue = 'sex', palette='rainbow')
# try hue = 'smoker'
```

###violinplot()

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill",data=tips, hue='sex', split=True)
```

###barplot()

```python
tips[tips['day']=='Thur']['total_bill'].mean()
```

```python
#sns.barplot(x='sex', y='total_bill',data=tips)
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", hue="sex",data=tips)
```

```python
tips[tips['day']=='Thur']['total_bill'].std()
```

```python
import numpy as np
#sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
sns.barplot(x="day", y="total_bill", data=tips, estimator=np.std)
```

###countplot()

```python
tips[tips['day']=='Sat']['total_bill'].count()
```

```python
sns.countplot(x="day", data=tips)
```

```python
sns.countplot(x="day", hue="sex",data=tips)
```

```python
# another example using size in the data
sns.countplot(x="size",data=tips)
# group of 2 is the most common!
#sns.countplot(x='size',hue='day',data=tips) # try this line of code
```
